```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
np.set_printoptions(suppress=True)


dataa = pd.read_csv('/Users/hinataboke/Desktop/Project2024_å¼‚å¸¸å€¼.csv', encoding='utf-8')
features = ['danceability', 'energy', 'acousticness', 'artist_popularity', 'valence']
Q1 = dataa[features].quantile(0.25)
Q3 = dataa[features].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
outliers = ((dataa[features] < lower_limit) | (dataa[features] > upper_limit)).any(axis=1)
dataaa = dataa[~outliers]

mean = dataaa[features].mean()
std = dataaa[features].std()

lower_limit = mean - 2 * std
upper_limit = mean + 2 * std
outliers_std = ((dataaa[features] < lower_limit) | (dataaa[features] > upper_limit)).any(axis=1)
data = dataaa[~outliers_std]
X = data[['danceability', 'energy', 'acousticness', 'artist_popularity', 'valence']]
y = data['track_pop']
model = LinearRegression()
model.fit(X, y)
weights = model.coef_
intercept = model.intercept_
print("weights:", weights)
print("interceptï¼š", intercept)

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 10))
axes = axes.flatten()
for i, feature in enumerate(X.columns[1:]):
    sns.regplot(x=X[feature], y=y, ax=axes[i], scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
    axes[i].set_title(f'{feature} vs Track Popularity')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Track Popularity')

plt.tight_layout()
plt.show()


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(X_train, y_train)

weights = model.coef_
features = X.columns
intercept = model.intercept_
weights_df = pd.DataFrame({'Feature': features, 'Weight': weights})


y_pred = model.predict(X_test)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test, y=y_pred, color='blue', alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--') 
plt.title('Actual vs. Predicted Popularity')
plt.xlabel('Actual Popularity')
plt.ylabel('Predicted Popularity')
plt.grid(True)
plt.show()
print(weights_df)
print("intercept:", intercept)
```

```python
data_2022 = pd.read_csv('Project2022.csv')
data_2023 = pd.read_csv('Project2023.csv')
data_2024 = pd.read_csv('Project2024.csv')

def analyze_yearly_data(data, year):
    print(f"\n=== Analyzing Data for {year} ===")
    variables = ['danceability', 'energy', 'acousticness', 'artist_popularity', 'valence', 'track_pop']
    if not all(var in data.columns for var in variables):
        raise ValueError(f"data {year} is missing necessary columns! Check the data structure")
    
 #Central limit theorem
    sample_size = 30  
    num_samples = 1000
    sample_means = []
    for _ in range(num_samples):
        sample = data['track_pop'].sample(sample_size, replace=True) 
        sample_means.append(sample.mean())

    plt.figure(figsize=(10, 6))
    sns.histplot(sample_means, kde=True, color='blue', bins=30)
    plt.title(f'Central Limit Theorem: Sample Means Distribution ({year})')
    plt.xlabel('Sample Mean of track_pop')
    plt.ylabel('Frequency')
    plt.show()

    mean_of_means = np.mean(sample_means)
    std_of_means = np.std(sample_means)
    print(f"Sample Means for {year}: Mean = {mean_of_means:.2f}, Std Dev = {std_of_means:.2f}")

#Chebyshev's inequality
    mean_track_pop = data['track_pop'].mean()
    std_track_pop = data['track_pop'].std()

    k_values = [1, 2, 3] 
    plt.figure(figsize=(10, 6))

    sns.histplot(data['track_pop'], kde=True, color='orange', bins=30, label='Actual Distribution')

    for k in k_values:
        lower_bound = mean_track_pop - k * std_track_pop
        upper_bound = mean_track_pop + k * std_track_pop
        prob = 1 - 1 / k**2  
        plt.axvline(x=lower_bound, color='green', linestyle='--', label=f'k={k}, Lower Bound')
        plt.axvline(x=upper_bound, color='red', linestyle='--', label=f'k={k}, Upper Bound')
        print(f"Chebyshev Inequality ({year}): At k={k}, at least {prob:.2%} of data lies within [{lower_bound:.2f}, {upper_bound:.2f}]")

    plt.title(f'Chebyshev Inequality: track_pop Distribution with Bounds ({year})')
    plt.xlabel('track_pop')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

#Analyze the relationship of track_pop to five independent variables
    plt.figure(figsize=(10, 6))
    correlation_matrix = data[variables].corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(f'Correlation Matrix of track_pop and Variables ({year})')
    plt.show()

    for var in ['danceability', 'energy', 'acousticness', 'artist_popularity', 'valence']:
        plt.figure(figsize=(8, 5))
        sns.scatterplot(x=data[var], y=data['track_pop'], alpha=0.6, color='purple')
        plt.title(f'Relationship between {var} and track_pop ({year})')
        plt.xlabel(var)
        plt.ylabel('track_pop')
        plt.show()
#Analyze data for 2022
analyze_yearly_data(data_2022, 2022)
```
